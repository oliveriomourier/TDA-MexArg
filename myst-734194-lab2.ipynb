### Swaps Metrics
*Using all data points, produce de following table (almost similar to Lab1)*

```python
import pandas as pd
import datetime
import os
import sys
import json
import numpy as np
```

```python
with open('uniswapv3_all_swaps_sqrtPriceX96.json', 'r') as archivo:
    json_data = json.load(archivo)
```

```python
price_wethusdc = []
for i in range(0,len(json_data)):
    data = json_data[i]
    price96 = float(data["sqrtPriceX96"])
    price1 = (price96 / (2**96)) ** 2
    decimals_weth = 10e18
    decimals_usdc = 10e6
    decimals_wethusdc = int(decimals_weth / decimals_usdc)
    price = decimals_wethusdc / price1
    price_wethusdc.append(price)
    i = i+1
```

```python
nt = []
for i in range(len(json_data)):
    t = json_data[i]['timestamp']
    nt.append(t)
    i = i+1
```

```python
volume_quote = []
for i in range(len(json_data)):
    t = json_data[i]['amount0']
    volume_quote.append(t)
    i = i+1
```

```python
volume_base = []
for i in range(len(json_data)):
    t = json_data[i]['amount1']
    volume_base.append(t)
    i = i+1
```

```python
df_swaps = pd.DataFrame({
                    'id': range(1, len(json_data) + 1),
                    'timestamp': pd.to_datetime((nt), unit='s'),
                    'weth/usdc': price_wethusdc,
                    'symbol_quote': 'USDC',
                    'volume_quote': (volume_quote),
                    'symbol_base': 'WETH',
                    'volume_base': (volume_base)
                    })
```

```python
df_swaps['volume_quote'] = pd.to_numeric(df_swaps['volume_quote'])
df_swaps['volume_base'] = pd.to_numeric(df_swaps['volume_base'])
```

```python
def swap(valor):
    if valor > 0:
        return 'buy'
    else:
        return 'sell'

df_swaps['swap'] = df_swaps['volume_base'].apply(swap)
```

```python
def swap_detail(valor):
    if valor > 0:
        return 'Swap USDC for WETH'
    else:
        return 'Swap WETH for USDC'

df_swaps['swap_detail'] = df_swaps['volume_base'].apply(swap_detail)
```

### Does the price follow a martingale process ?

**Pt = E [Pt+1] ≡ E [Pt+1] = Pt**

```python
price_mg = [ ]
for i in range(len(json_data)):
    precios = json_data[i]['pool']["token0Price"]
    price_mg.append(precios)
    i = i+1
price_mg = list(map(float, price_mg))
```

```python
price_e = sum(price_mg) / len(price_mg)
```

```python
if price_e == price_mg[-1]:
    print("The price follows a martingale process")
else:
    print("The price does not follow a martingale process")
```
